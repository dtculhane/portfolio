```python
import pandas as pd

megamill = pd.read_csv('Lottery_Mega_Millions_Winning_Numbers__Beginning_2002_20240421.csv')
nypick10 = pd.read_csv('Lottery_Pick_10_Winning_Numbers__Beginning_1987_20240421.csv')
powball = pd.read_csv('Lottery_Powerball_Winning_Numbers__Beginning_2010_20240421.csv')
```

*CONCATENATE SECTIONS TOGETHER*

```python
pow_nums = pd.lreshape(powball,
           {'Number' : ['WinNum 1', 'WinNum 2', 'WinNum 3', 'WinNum 4', 'WinNum 5', 'WinNum 6']}).drop(columns=['Multiplier'])
pow_mult = powball.get(['Draw Date', 'Multiplier'])
```

```python
nypick10_nums = pd.lreshape(nypick10,
           {'Number' : ['WinNum 1', 'WinNum 2', 'WinNum 3', 'WinNum 4', 'WinNum 5', 'WinNum 6',
                       'WinNum 7', 'WinNum 8', 'WinNum 9', 'WinNum 10','WinNum 11','WinNum 12',
                       'WinNum 13','WinNum 14','WinNum 15','WinNum 16','WinNum 17','WinNum 18',
                       'WinNum 19','WinNum 20']})
```

```python
mega_nums = pd.lreshape(megamill,
           {'Number' : ['WinNum 1', 'WinNum 2', 'WinNum 3', 'WinNum 4', 'WinNum 5']}).drop(columns=['Mega Ball', 'Multiplier'])
mega_mult = megamill.get(['Draw Date', 'Multiplier'])
mega_balls = megamill.get(['Draw Date', 'Mega Ball'])
```

With the data for each lottery broken down, we can write each to a spreadsheet for use in PowerBI.

```python
# Writing the Powerball Data
pow_nums.to_excel('powerball numbers.xlsx', index=False, sheet_name='Powerball Numbers')
pow_mult.to_excel('powerball multipliers.xlsx', index=False, sheet_name='Powerball Multipliers')

# Writing the Mega Millions Data
mega_nums.to_excel('mega millions numbers.xlsx', index=False, sheet_name='Mega Millions Numbers')
mega_mult.to_excel('mega millions multipliers.xlsx', index=False, sheet_name='Mega Millions Multipliers')
mega_balls.to_excel('mega millions mega balls.xlsx', index=False, sheet_name='Mega Millions Mega Ball')

# Writing the NY Pick 10 Data
nypick10_nums.to_excel('ny pick 10 numbers.xlsx', index=False, sheet_name='NY Pick 10 Numbers')
```
